```python
import cv2
import numpy as np

# Read the image
image_path = "Downloads/i/coconut/10.jpg"
image = cv2.imread(image_path)

# Get the center of the image
center = (image.shape[1] // 2, image.shape[0] // 2)

# Create a mask with a circular shape
mask = np.zeros_like(image)
cv2.circle(mask, center, min(center), (255, 255, 255), -1, cv2.LINE_AA)

# Create a white background image
background = np.ones_like(image) * 255

# Apply the mask to the images
result = cv2.bitwise_and(image, mask)
background = cv2.bitwise_and(background, cv2.bitwise_not(mask))

# Combine the cropped circular region with the white background
result = cv2.add(result, background)

# Save the result
cv2.imwrite("Downloads/i/coconut/10_cropped.jpg", result)

cv2.namedWindow("Image with White Background", cv2.WINDOW_NORMAL)
cv2.resizeWindow("Image with White Background", 500, 400)

# Display the result
cv2.imshow('Image with White Background', result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

image=cv2.imread("Downloads/i/coconut/cropped_patch1.jpg")

_,binary_image=cv2.threshold(image,50,255,cv2.THRESH_BINARY)

kernel=np.ones((5,5),dtype=np.uint8)
eroded_image=cv2.erode(binary_image, kernel, iterations=1)

cv2.namedWindow("gray_image", cv2.WINDOW_NORMAL)
cv2.resizeWindow("gray_image", 500, 400)
cv2.imshow("gray_image",image)

cv2.namedWindow("thresh_image", cv2.WINDOW_NORMAL)
cv2.resizeWindow("thresh_image", 500, 400)
cv2.imshow("thresh_image",binary_image)

cv2.namedWindow("eroded_image", cv2.WINDOW_NORMAL)
cv2.resizeWindow("eroded_image", 500, 400)
cv2.imshow("eroded_image",eroded_image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
